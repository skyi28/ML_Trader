```python
ptd = PrepareTrainingData()
data = ptd.load_data('BTCUSD')
data = data.pipe(ptd.remove_na).pipe(ptd.remove_timestamp).pipe(ptd.add_return).pipe(ptd.add_direction, remove_zeros=True)
features, target = ptd.create_features_and_targets(data)
features.tail()
```

```python
xgb_model_class = XGBoostModel(0)
xgb_model = xgb_model_class.create_model(mode='direction')
xgb_model = xgb_model_class.train(xgb_model, features, target)
```
